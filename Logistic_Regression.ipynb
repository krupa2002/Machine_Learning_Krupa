# Logistic Regression

Logistic regression is a statistical method used to model the relationship between a binary outcome and one or more predictor variables by estimating the probability of the outcome occurring. It's particularly useful for classification tasks where the dependent variable is categorical.

Logistic regression predicts if something will happen (yes/no) by looking at other related factors.

Let's say you want to predict whether a student will pass or fail an exam based on the number of hours they study. You collect data on hours studied (predictor variable) and whether they passed or failed (outcome variable). Logistic regression helps you estimate the probability of passing based on the hours studied.

Logistic regression is like drawing a curve on a graph instead of a straight line. It helps us predict if something will happen or not (like pass or fail), based on factors like study hours.

### 3 Main reasons why we use logistic regression

1. **Binary Outcome Prediction**: Logistic regression is well-suited for predicting binary outcomes, such as pass/fail, yes/no, or win/lose, making it ideal for classification tasks.

2. **Probability Estimation**: It provides probabilities of belonging to each category, allowing us to understand the likelihood of different outcomes based on predictor variables.

3. **Interpretability**: Logistic regression coefficients can be interpreted as the effect of predictor variables on the log-odds of the outcome, providing insights into the relationship between predictors and the likelihood of the outcome.

### When we should  logistic regression

1. **Binary Outcome**: You have a binary outcome variable, like pass/fail, yes/no, or win/lose.

2. **Linear Relationship**: There is a linear relationship between the predictor variables and the log-odds of the outcome.

3. **Assumption of Independence**: Observations are independent of each other, meaning one observation doesn't influence another.

4. **Few Predictors**: You have relatively few predictor variables compared to the number of observations, as logistic regression may struggle with overfitting when the number of predictors is large relative to the number of observations.

5. **Interpretability**: You need a model that provides interpretable coefficients for understanding the relationship between predictors and the outcome.

### Why Logistic Regression comes under classfication and not regression?

1. **Categorical Outcome**: Logistic regression is specifically designed for modeling binary or multinomial outcomes, where the outcome variable is categorical and not continuous. It predicts the probability of belonging to a particular category (e.g., pass/fail, yes/no) rather than predicting a numerical value.

2. **Model Output**: In logistic regression, the output is the probability of the event occurring (e.g., probability of passing an exam), which is then used to classify observations into categories based on a threshold. In contrast, regression models typically predict a continuous value.

3. **Evaluation Metrics**: Classification models, including logistic regression, are evaluated using metrics like accuracy, precision, recall, and F1-score, which are suited for assessing the performance of models in predicting categorical outcomes. Regression models, on the other hand, are evaluated using metrics like mean squared error (MSE) or R-squared, which are appropriate for continuous predictions.


### Statistical Analysis for Logistic Regression

1. **Model**: pi = 1 / (1 + exp(-(0 + 1*X1i + 2*X2i + ... + k*Xki)))

- `pi` represents the predicted probability of success for observation `i`.
- `0` is the intercept term.
- `1`, `2`, ..., `k` are the coefficients for the predictor variables `X1`, `X2`, ..., `Xk`.
- `X1i`, `X2i`, ..., `Xki` are the values of the predictor variables for observation `i`.
- `exp()` denotes the exponential function.

### But how the logistic regression finds out which is the best fit line?

Logistic regression finds the best-fit parameters (coefficients) by maximizing the likelihood function, which measures how well the model predicts the observed binary outcomes. This optimization process iteratively adjusts the parameters until convergence, resulting in coefficients that describe the relationship between predictor variables and the log-odds of the binary outcome.

### Types of Logistic regression

1. Binary Logistic Regression
2. Multinomial Logistic Regression
3. Ordinal Logistic Regression
4. Regularized Logistic Regression

###Steps to implement Logistic Regression

1. **Data Preparation**:
   - Gather and preprocess the dataset, handling missing values and encoding categorical variables if necessary.
   - Split the data into training and testing sets for model evaluation.

2. **Model Initialization**:
   - Initialize the logistic regression model.

3. **Model Training**:
   - Train the logistic regression model on the training data.
   - Optimize the model parameters (coefficients) using an optimization algorithm like gradient descent or its variants.
   
4. **Model Evaluation**:
   - Evaluate the trained model's performance on the testing data.
   - Calculate evaluation metrics such as accuracy, precision, recall, F1-score, and ROC-AUC.

5. **Model Interpretation**:
   - Interpret the coefficients of the logistic regression model to understand the relationship between predictor variables and the probability of the outcome.

6. **Prediction**:
   - Use the trained logistic regression model to make predictions on new data.
   - Compute the predicted probabilities of the outcome for each observation.

7. **Model Tuning**:
   - Perform hyperparameter tuning, such as adjusting regularization strength or feature selection, to optimize model performance.

8. **Deployment**:
   - Deploy the trained logistic regression model into production for making predictions on unseen data.

### Loss function Cost function

The loss function measures the disparity between the actual binary outcomes and the predicted probabilities assigned by the model. The aim is to minimize this disparity, indicating a better fit of the model to the data.

1. **Binary Cross-Entropy Loss (Log Loss)**:
   - Measures the discrepancy between the true binary outcomes (0 or 1) and the predicted probabilities. It penalizes incorrect predictions more heavily when the true label is 1.
   
2. **Categorical Cross-Entropy Loss**:
   - Generalizes binary cross-entropy to handle multiple classes in a classification problem. It compares the predicted probabilities of each class to the true one-hot encoded labels.
   
3. **Hinge Loss**:
   - Used in support vector machines (SVMs) and linear classifiers. It measures the maximum distance between the predicted scores and the correct scores, with a margin of 1.


### Evaluation Metrics

Evaluation metrics for logistic regression assess the performance of the model in predicting binary outcomes.

1. **Accuracy**: The proportion of correctly classified instances out of the total instances.

2. **Precision**: The proportion of true positive predictions out of all positive predictions made by the model. It measures the accuracy of positive predictions.

3. **Recall (Sensitivity)**: The proportion of true positive predictions out of all actual positive instances in the data. It measures the ability of the model to identify positive instances correctly.

4. **F1-score**: The harmonic mean of precision and recall. It provides a balance between precision and recall.

5. **ROC-AUC (Receiver Operating Characteristic - Area Under the Curve)**: A metric that measures the area under the ROC curve. It represents the trade-off between true positive rate (sensitivity) and false positive rate across different probability thresholds.

6. **Confusion Matrix**: A table that summarizes the performance of a classification algorithm. It shows the counts of true positive, true negative, false positive, and false negative predictions.


### Application of Logistic Regression

1. **Healthcare**: Predicting the likelihood of a patient developing a specific disease based on their medical history, genetic factors, and lifestyle choices.

2. **Marketing**: Predicting customer churn (whether a customer will stop using a product or service) based on their behavior, demographics, and purchasing history.

3. **Finance**: Assessing credit risk by predicting the probability of default on loans using financial indicators and borrower information.
